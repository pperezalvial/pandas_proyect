```python

import pandas as pd

pd.options.mode.chained_assignment = None  # default='warn'
import numpy as np
import warnings
from pandas.tseries.offsets import MonthBegin
with warnings.catch_warnings():
    warnings.filterwarnings('ignore', r'elementwise comparison failed; returning scalar instead, but in the future will perform elementwise comparison(.*)')


# SUSPENSIONES PARTE 1

archivo = r'Z:\CONTROL DE GESTION\SEGUIMIENTO AGENDA Y PRODUCCION\Seguimiento programación médicos.xlsx'

#bASE Urgencias 
df = pd.read_excel(archivo, sheet_name='BD_SEGUIMIENTO') # modificar hoja por fecha a consultar
df1 = df.rename(columns={'META % ALTAS':'META % ALTAS'})





#df1['META % ALTAS'] = df1['META % ALTAS'].replace("%","",inplace=True).astype(float)
#df1['META_ALTAS'] = df1['META_ALTAS'].apply(lambda x: float(x.split()[0].replace('%', '')))
df1['META % ALTAS'] = df1['META % ALTAS'].replace("%","") 
df1['META % ALTAS'] = df1['META % ALTAS'].replace("-","")
df1['META % ALTAS'] = df1['META % ALTAS'].replace('','')
df1['META % ALTAS'] = df1['META % ALTAS'].replace(" ","")
df1['META % ALTAS'] = pd.to_numeric(df1['META % ALTAS'], errors='coerce')
#np.df1['META % ALTAS'] = np.where(df1['META % ALTAS']== '',0)

df1
```

```python
df1['TOTAL BLOQUEO'] = (df1['CUPO CONTROL BLOQUEADO'] + df1['CUPO NUEVO BLOQUEADO']) 
df1['PERDIDOS'] = (df1['PERDIDO NUEVO'] + df1['PERDIDO CONTROL']) 
df1= df1[(df1["AÑO"]==2020)]

raao = df1.groupby(["SERVICIO","ESPECIALIDAD","SUBESPECIALIDAD","MES"]).agg(
                                  {
                                      'HORAS SEMANALES PROGRAMADAS': 'sum',
                                      'RENDIMIENTO NUEVO': 'mean',#sum
                                      'RENDIMIENTO CONTROL': 'mean',#sum
                                      'CONSULTAS MENSUALES PROGRAMADAS':'sum',
                                         'CONSULTA TOTAL INICIAL':'sum',
                                              'TOTAL BLOQUEO' : 'sum',
                                       'OFERTA TOTAL FINAL': 'sum',
                                
                                       'CONSULTAS NUEVAS REALIZADAS' : 'sum',
                                       'CONSULTAS CONTROL REALIZADAS' : 'sum',
                                      'TOTAL CONSULTAS REALIZADAS' : 'sum',
                                        'NSP NUEVOS' : 'sum',
                                       'NSP CONTROL' : 'sum',
                                       'TOTAL NSP' : 'sum',
                                      'ALTAS' : 'sum',
                                      'META % NUEVA': 'mean',#mean
                                     'META % ALTAS': 'mean',
                                      'SOBRECUPO': 'sum',
                                     'PERDIDOS': 'sum'
                                                                                  
                                                                           
                                      
                                     
                                      
                                  }).reset_index()





raao['% CONSULTAS NUEVAS REALIZADAS'] = ((raao['CONSULTAS NUEVAS REALIZADAS'] / raao['TOTAL CONSULTAS REALIZADAS'])*100)
raao['% ALTAS'] = ((raao['ALTAS'] / raao['TOTAL CONSULTAS REALIZADAS'])*100)

raao['% TOTAL CUPOS BLOQUEADOS'] = ((raao['TOTAL BLOQUEO'] / raao[ 'CONSULTA TOTAL INICIAL'])*100)
raao['% TOTAL NSP'] = ((raao['TOTAL NSP'] / raao['OFERTA TOTAL FINAL'])*100)
raao['% CUMPLIMIENTO CONSULTAS'] = ((raao['TOTAL CONSULTAS REALIZADAS'] / raao['OFERTA TOTAL FINAL'])*100)
raao['% SOBRECUPO'] = ((raao['SOBRECUPO'] / raao['TOTAL CONSULTAS REALIZADAS']))
raao['% PERDIDOS'] = (raao['PERDIDOS'] / raao['OFERTA TOTAL FINAL']*100)



raao.set_index('SERVICIO', inplace=True)
df22 = raao.rename(columns={'HORAS SEMANALES PROGRAMADAS':'HRS_SEMANALES_PROGRA','RENDIMIENTO NUEVO':'RENDIMIENTO_NUEVO','RENDIMIENTO CONTROL':'RENDIMIENTO_CONTROL',
                           'CONSULTAS MENSUALES PROGRAMADAS':'CONSULTAS_MENSUALES_PROGRA','CONSULTA TOTAL INICIAL':'CONSULTA_TOTAL_INICIAL','TOTAL BLOQUEO':'TOTAL_BLOQUEO',
                           'OFERTA TOTAL FINAL':'OFER_TOTAL_FINAL','CONSULTAS NUEVAS REALIZADAS':'CONSULTAS_NUEVAS_REALIZADAS','CONSULTAS CONTROL REALIZADAS':
                            'CONSULTAS_CONTROL_REALIZADAS', 'TOTAL CONSULTAS REALIZADAS':'TOTAL_CONSULTAS_REALIZADAS','NSP NUEVOS':'NSP_NUEVOS','NSP CONTROL':'NSP_CONTROL'
                            ,'TOTAL NSP':'TOTAL_NSP','META % NUEVA':'META_PORC_NUEVA','META % ALTAS':'META_PORC_ALTAS','% CONSULTAS NUEVAS REALIZADAS':'PORC_CONSULTAS_NUEVAS_REALIZADAS',
                            '% ALTAS':'PORC_ALTAS','% TOTAL CUPOS BLOQUEADOS':'PORC_TOTAL_CUPOS_BLOQ','% TOTAL NSP':'PORC_TOTAL_NSP','% CUMPLIMIENTO CONSULTAS':'PORC_CUMPLIMIENTO_CONSULTAS',
                            '% SOBRECUPO':'PORC_SOBRECUPO','% PERDIDOS':'PORC_PERDIDOS'})
#raao = raao.assign(total=raao.sum(6))

df23=df22.replace([np.inf, -np.inf])

#raao = raao.stack(level='HORAS SEMANALES PROGRAMADAS')
#raao$NumPatatasTotal = rowSums (raao[ , 5:28])
df23
#raao.tail(1)
```
