```python
import pandas as pd

# DICCIONARIO DE DATOS DE TIPO SESSION
archivo = 'C:/Users/manuel.caceres.HBL/Documents/DB/AGENDAS MEDICAS/DICCIONARIO_AGENDA_CORTA.xlsx'


df3 = pd.read_excel(archivo, sheet_name='Hoja1', usecols=['Especialidad', 'SERVICIO' , 'ESP','SUBESPECIALIDAD','SM-OS-OP','DETALLE'
])

print(df3)
```

```python

indexed_df3= df3.set_index('Especialidad') 
dfNew2 = pd.merge( dfNew1,  indexed_df3 , how = 'left', on='Especialidad' ) 
dfNew2['OFERTA INICIAL NUEVOS'] = np.where(dfNew2['resumen_tipo']=='nueva', 1, 0)
dfNew2['OFERTA INICIAL CONTROL'] = np.where(dfNew2['resumen_tipo']=='repetida', 1, 0) # condicion para egregar flag 1 o 0 cegun corresponda y se cumnpla la comndicion
dfNew2['OFERTA TOTAL INICIAL'] = dfNew2[['OFERTA INICIAL NUEVOS','OFERTA INICIAL CONTROL']].sum(axis=1)# suma los campos indicados

#crear condicion con dos columnas BLOQ NUEVOS
conditions = [
       (dfNew2['resumen_tipo'] == 'nueva') & (dfNew2['Bloqueo'] == 'BLOQUEADA')
      ]
choices = [1]
dfNew2['BLOQ NUEVOS'] = np.select(conditions, choices, default=0) 

#crear condicion con dos columnas BLOQ TOTAL
conditions = [
       (dfNew2['resumen_tipo'] == 'repetida') & (dfNew2['Bloqueo'] == 'BLOQUEADA')
      ]
choices = [1]
dfNew2['BLOQ CONTROL'] = np.select(conditions, choices, default=0) #crear condicion con dos columnas

dfNew2['BLOQ TOTAL'] = dfNew2[['BLOQ NUEVOS','BLOQ CONTROL']].sum(axis=1)
a = dfNew2['BLOQ TOTAL']

dfNew2['OFERTA FINAL'] = dfNew2['OFERTA TOTAL INICIAL'] - dfNew2['BLOQ TOTAL']




dfNew2
```

```python
import pandas as pd

# DICCIONARIO DE DATOS DE TIPO SESSION
archivo = 'C:/Users/manuel.caceres.HBL/Documents/DB/Diccionario_contrato.xlsx'


di_con = pd.read_excel(archivo, sheet_name='Hoja1', usecols=['Tipo_Contrato', 'TIPO CONTRATO' ])


print(di_con)
```

```python
import datetime
from datetime import datetime

indexed_df4= di_con.set_index('Tipo_Contrato') 
dfNew3 = pd.merge( dfNew2,  indexed_df4 , how = 'left', on='Tipo_Contrato' ) 


now = datetime.now()
now.year
now.month - 1
dfNew3["AÑO"] = now.year

dfNew3["MES" ] = now.month - 1
#crear condicion con dos columnas SOBRECUPO
conditions = [
       (dfNew3['Bloqueo'] != 'BLOQUEADA' ) & (dfNew3['Motivo_Sobrecupo'] == '' )
      ]
choices = [1]
dfNew3['SOBRECUPO'] = np.select(conditions, choices, default=0) 

#crear condicion con 4 columnas PERDIDO NUEVO
conditions = [
    (dfNew3['resumen_tipo'] == 'nueva') &
    (dfNew3['Estado_Citacion'] == 'Perdido')&
    (dfNew3['Bloqueo'] != 'BLOQUEADA' ) ]
choices = [1]

dfNew3['PERDIDO NUEVO'] = np.select(conditions, choices, default=0)

conditions = [
    (dfNew3['Bloqueo'] != 'BLOQUEADA' )&
    (dfNew3['resumen_tipo'] == 'repetida') &
    (dfNew3['Estado_Citacion'] == 'Perdido')
     ]
choices = [1]

dfNew3['PERDIDO CONTROL'] = np.select(conditions, choices, default=0) 

dfNew3 = dfNew3.rename(columns={'Fecha_Citacion_DATE':'fecha', 
                               'Especialidad':'Especialidad_local', 'Bloqueo':'Motivo Bloqueo', 'SUBESPECIALIDAD':'SUBESP'})


dfNew3 = dfNew3.rename(columns={'SERVICIO':'SERV', 'ESP': 'ESPECIALIDAD' })





df4 = dfNew3[['AÑO','MES','fecha','SERV','SM-OS-OP', 'DETALLE','ESPECIALIDAD', 'Especialidad_local','Profesional','Tipo_Profesional','resumen_tipo','Motivo Bloqueo','Motivo_Sobrecupo','Estado_Citacion',
              'resumen para perdidos','OFERTA INICIAL NUEVOS','OFERTA INICIAL CONTROL','OFERTA TOTAL INICIAL','BLOQ NUEVOS','BLOQ CONTROL','BLOQ TOTAL','OFERTA FINAL',
             'SUBESP','TIPO CONTRATO','SOBRECUPO','PERDIDO NUEVO','PERDIDO CONTROL']]
df4
```

```python

#genera tabla dinamica sumando todos los flag

agrupado  = df4.groupby(['MES', 'SM-OS-OP','SERV','ESPECIALIDAD','SUBESP','Especialidad_local','Profesional','TIPO CONTRATO',
                         'Tipo_Profesional','DETALLE']).agg(
                                  {
                                      'OFERTA INICIAL NUEVOS' : 'sum',
                                    'OFERTA INICIAL CONTROL' : 'sum',  
                                   'OFERTA TOTAL INICIAL' : 'sum',
                                   'BLOQ NUEVOS': 'sum', 
                                   'BLOQ CONTROL':'sum', 
                                   'BLOQ TOTAL':'sum',
                                   'OFERTA FINAL':'sum',
                                    'SOBRECUPO':'sum', 
                                   'PERDIDO NUEVO':'sum',
                                   'PERDIDO CONTROL':'sum'
                                  }).reset_index()


df4['Suma de perdidos tot'] = df4[['PERDIDO NUEVO','PERDIDO CONTROL']].sum(axis=1)

#result = pd.merge(df4, agrupado, on=['MES', 'SM-OS-OP','SERV'])[['BLOQ NUEVOS', 'BLOQ CONTROL', 'BLOQ TOTAL', 'OFERTA FINAL']]
agrupado2  = df4.groupby([ 'AÑO','MES', 'SERV','ESPECIALIDAD','SUBESP','Especialidad_local','Profesional','Tipo_Profesional','TIPO CONTRATO','resumen para perdidos']).agg(
                                  {
                                    'Suma de perdidos tot' : 'sum'
                                  }).reset_index()
```

```python
#ultimo paso, exporta la tabla agenda cortas, separando cada dataframe en hojas diferentes, a fin de mostrar disntintas vistas.

from datetime import datetime
import time


TodaysDate = time.strftime("%d-%m-%Y")
#se crea una clase tiempo

#Se asigna la clase al objeto csvfilename para luego agregar la fecha al export
exceldate = TodaysDate +".xlsx"

from pandas import ExcelWriter
writer = ExcelWriter(r'C:/Users/manuel.caceres.HBL/Documents/DB/AGENDAS MEDICAS/td_agendas_cortas' + exceldate)
df4.to_excel(writer, 'BD')
agrupado.to_excel(writer, 'TD')
agrupado2.to_excel(writer, 'PERDIDOS')
writer.save()
```

```python

## HTML y EXCEL ##

# Importamos la librería pandas:
import pandas as pd

# Guardamos la url de una web que contenga una tabla con datos dentro de una variable:
url = 'https://es.wikipedia.org/wiki/Anexo:Finales_de_la_Copa_Mundial_de_F%C3%BAtbol'

# Creamos un dataframe con los datos de la tabla de esa web:
dataframe = pd.io.html.read_html(url)

# Nos quedamos con la primera tabla de la web:
dataframeFutbol = dataframe[0]

# En el dataframe, los nombres de las columnas se encuentran en realidad en la primera fila de la tabla, y no en la cabecera, así que...
# Creamos un diccionario con la primera fila de datos y lo asignamos al dataframe como nuevos nombres de columnas:
dataframeFutbol = dataframeFutbol.rename(columns = dict(dataframeFutbol.loc[0]))

# Y eliminamos la primera fila de datos, que ya hemos colocado antes como nombres de las columnas:
dataframeFutbol = dataframeFutbol.drop(0)
dataframeFutbol
```
